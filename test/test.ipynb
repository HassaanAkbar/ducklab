```python

import pandas as pd

import glob
```

```python

cols = set()

for fil in glob.glob("D:/data/Blackblaze_Hard_Drive_2023/data_Q*_2023/*.csv"):
    cols = cols.union(pd.read_csv(fil).columns)
    print(fil)

    
```

```python

for fil in glob.glob("D:/data/Blackblaze_Hard_Drive_2023/data_Q*_2023/*.csv"):
    df = pd.read_csv(fil)
    df = df.reindex(columns=list(cols))
    df.to_csv(fil, index=False)
```

```python

import polars as pl

data = pl.scan_csv("D:/data/Blackblaze_Hard_Drive_2023/data_Q*_2023/*.csv")

ctx = pl.SQLContext()
ctx.register("data", data)

result = ctx.execute("select date, count(*) from data group by date")
print(result.collect())
```
